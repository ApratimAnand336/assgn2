```python
#why scaling before pca?
#this is so because the pca is based on the original varience of the different features, and hence the different varience can create bias towards or against any feature and also its for mean centering
#hence it is required to scale the features before pca
```

```python
#to apply pca over the data without using sklearn direcctly as it is not mentioned to use
from sklearn.preprocessing import StandardScaler

#so first scale the data
scaler=StandardScaler()
data_for_pca=data.drop(['ocean_proximity','median_house_value'],axis=1).copy()
scaled_data=scaler.fit_transform(data_for_pca)
```

```python
scaled_df=pd.DataFrame(scaled_data,columns=data.drop(['ocean_proximity','median_house_value'],axis=1).columns)
```

```python
scaled_df.head()
```

```python
cov_mat=scaled_df.cov()
cov_mat
```

```python
#now to get the eigen value and eigen vectors, through which we will find th etwo with the highest eigen values as out pc1 and pc2

e_val,e_vec=np.linalg.eig(cov_mat)
```

```python
e_val
```

```python
e_vec
```

```python
pc_vect=e_vec[0:2,:]
```

```python
pc_vect
```

```python
print(pc_vect.shape,scaled_data.shape)
```

```python
#now transforming the data to make it have only two feature and label
data_transform=np.dot(scaled_data,pc_vect.T)
transformed_df=pd.DataFrame(data_transform,columns=["PC1","PC2"])
transformed_df['median_house_value']=data['median_house_value']
```

```python
transformed_df.head()
```

```python
plt.scatter(transformed_df['PC1'],transformed_df["PC2"],c=transformed_df["median_house_value"],alpha=0.3)
plt.ylabel('PC1')
plt.xlabel('PC2')
plt.colorbar()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
#we need to encode the column with string rather tahn numerical data i.e. ocean proximity
#then we will train test split the data
encoder=OneHotEncoder(sparse_output=False)
encoded_data=encoder.fit_transform(data[['ocean_proximity']])
col_name=encoder.get_feature_names_out(['ocean_proximity'])
encoded_df=pd.DataFrame(encoded_data,columns=list(col_name))
# encoded_df.head()

encoded_data=pd.concat([data.drop(['ocean_proximity','latitude','longitude'],axis=1),encoded_df],axis=1)
encoded_data.head()
```

```python
encoded_data.dropna(axis=0,inplace=True)
x=encoded_data.drop('median_house_value',axis=1).copy()
y=encoded_data['median_house_value']
# y=pd.DataFrame(encoded_data['median_house_value'],columns=['median_house_value'])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predicted=model.predict(x_test)
```

```python
r2=r2_score(y_true=y_test,y_pred=y_predicted)
mae=mean_absolute_error(y_true=y_test,y_pred=y_predicted)
mse=mean_squared_error(y_true=y_test,y_pred=y_predicted)
```
